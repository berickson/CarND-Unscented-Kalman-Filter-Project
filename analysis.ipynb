```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('output/out1.tsv',delimiter='\t')
```

```python
df
```

```python
fig = plt.figure(dpi=200)
fig.set_size_inches(h=15,w=15)

plt.title('location')
plt.xlabel('px_true')
plt.ylabel('py_true')
plt.plot(df.px_measured, df.py_measured, color='b')
plt.plot(df.px_true, df.py_true, color='g')

fig = plt.figure(dpi=200)
fig.set_size_inches(h=15,w=15)

plt.title('velocity')
plt.xlabel('vx_true')
plt.ylabel('vy_true')
plt.plot(df.vx_true, df.vy_true);
```
