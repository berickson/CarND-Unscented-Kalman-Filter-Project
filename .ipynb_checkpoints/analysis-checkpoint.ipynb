```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('output/out2.txt',delimiter='\t')
```

```python
plt.figure()
plt.title('location')
plt.xlabel('px_true')
plt.ylabel('py_true')
plt.plot(df.px_true, df.py_true)
plt.figure()

plt.title('velocity')
plt.xlabel('vx_true')
plt.ylabel('vy_true')
plt.plot(df.vx_true, df.vy_true);
```
